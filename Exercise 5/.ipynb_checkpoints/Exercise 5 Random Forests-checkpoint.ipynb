1. Install MATLAB Python API
    (http://de.mathworks.com/help/matlab/matlab_external/install-the-matlab-engine-for-python.html)
2. Start MATLAB (Working directory in RandomForestClassification folder)
3. Run "matlab.engine.shareEngine" in MATLAB
4. Run python script:

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matlab.engine
```

```python
eng = matlab.engine.connect_matlab()
```

Create 2D Gaussian data

```python
xtrain, ytrain, xtest, ytest = eng.gaussian_data_2d(nargout=4)
```

Study the performance (i.e. the error rate) of 1 tree on both the training set and the test set when the depth increases (depth varying between 2 and 20). What do you observe?

```python
error = []
for depth in range(2,21):
    error_train, error_test = eng.train_forest(xtrain, ytrain, xtest, ytest,1.0,float(depth),10.0,2.0,'random',nargout=2)
    error.append([error_train,error_test])
```

```python
plt.xlabel('Tree depth'); plt.ylabel('Error')
plt.plot(range(2,21),error)
plt.show()
```

Study the performance of the random forest classifier when the number of tree increases.

```python
error = []
for trees in range(1,40):
    error_train, error_test = eng.train_forest(xtrain, ytrain, xtest, ytest,float(trees),10.0,1.0,2.0,'random',nargout=2)
    error.append([error_train,error_test])
```

```python
plt.xlabel('Number of Trees');plt.ylabel('Error')
plt.plot(range(1,40),error)
plt.show()
```

High dimensional data (Relevant features = 20)

```python
xtrain, ytrain, xtest, ytest = eng.high_dimensional_data(20.0, nargout=4)
error = []
for tries in range(1,100):
    error_train, error_test = eng.train_forest(xtrain, ytrain, xtest, ytest,5.0,5.0,float(tries),100.0,'binary',nargout=2)
    error.append([error_train,error_test])
```

```python
plt.xlabel('Number of Tries');plt.ylabel('Error')
plt.plot(range(1,100),error)
plt.show()
```
